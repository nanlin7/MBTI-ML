```python
import pandas as pd
import re
from collections import Counter
```

```python
mbti = pd.read_csv('data/mbti_1.csv') 
mbti_features = mbti.copy()
mbti_features.drop(columns=['posts'],inplace=True)
```

# Length related features 
### Average length

```python
def avg_length(post): 
    posts = post.split("|||")
    post_length = 0
    for item in posts:
        post_length += len(item)
    return post_length/len(posts)
```

### Number of posts

```python
def num_posts(post):
    posts = post.split("|||")
    return len(posts)
```

### Number of long posts

```python
def num_long_posts(post):
    return len(re.findall(r'\.\.\.\|\|\|', post))
```

### Number of emojies
We should probably modify the text_style_emojies to include more.

```python
def find_emoji(text):
    # REMOVE LATER ON
    text = text.lower()
    
    text = re.sub(r'\|\|\|', ' ', text)
    
    slack_style_emojies = re.findall(r':[\w\d]+(\-[\w\d]+)?:', text)
    text_style_emojies = re.findall(r':[\-|\s]?[d|\)|\(|p]', text)
    
    return slack_style_emojies + text_style_emojies

# def most_common_emoji(text):
#     emoji_count = Counter(find_emoji(text))
#     if len(emoji_count) == 0:
#         return None    
#     return emoji_count.most_common(1)[0][0]
```

#### Test for emoji functions

```python
def test_find_emoji():
    assert len(find_emoji(":cool: hi :ok:")) == 2
    assert len(find_emoji(":) :D :P :( :-) hi")) == 5
    assert len(find_emoji("hello hi")) == 0
    assert len(find_emoji(":suspicion::hampster::tranquillity::rolldeyes::biggrin:")) == 5

test_find_emoji()
```

```python
# Adding the features
mbti_features['avg_post_length'] = mbti['posts'].map(avg_length)
mbti_features['number_of_long_posts'] = mbti['posts'].map(num_long_posts)
mbti_features['number_of_posts'] = mbti['posts'].map(num_posts)
mbti_features['number_of_emojies'] = mbti['posts'].map(lambda x: len(find_emoji(x)))
```
